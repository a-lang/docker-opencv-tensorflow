# Template matching

```python
import numpy as np
import imutils
import cv2

# 資料視覺化模組
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 不要顯示惱人的 warning 訊息
import warnings
warnings.filterwarnings('ignore')

# 在 Notebook 裡可即時繪圖
%matplotlib inline  

plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams.update({'font.size':13})
```

## source_01.jpg

```python
# load the source and template image
source = cv2.imread("source_01.jpg")
template = cv2.imread("template.jpg")
(tempH, tempW) = template.shape[:2]

# find the template in the source image
result = cv2.matchTemplate(source, template, cv2.TM_CCOEFF)
(minVal, maxVal, minLoc, (x, y)) = cv2.minMaxLoc(result)

# draw the bounding box on the source image
cv2.rectangle(source, (x, y), (x + tempW, y + tempH), (0, 255, 0), 2)

# show the images
plt.title("Source")
plt.imshow(cv2.cvtColor(source, cv2.COLOR_BGR2RGB))
plt.show()

plt.title("Template")
plt.imshow(cv2.cvtColor(template, cv2.COLOR_BGR2RGB))
plt.show()
```
