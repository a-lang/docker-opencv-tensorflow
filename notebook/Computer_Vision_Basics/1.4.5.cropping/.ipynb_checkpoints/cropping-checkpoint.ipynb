# Cropping

```python
import numpy as np
import imutils
import cv2

# 資料視覺化模組
import matplotlib.pyplot as plt

# 不要顯示惱人的 warning 訊息
import warnings
warnings.filterwarnings('ignore')

# 在 Notebook 裡可即時繪圖
%matplotlib inline  

plt.rcParams['figure.figsize'] = [15, 7]
plt.rcParams.update({'font.size':13})
```

```python
# load the image and show it
image = cv2.imread("../input/florida_trip.png")

plt.title("Original")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# cropping an image is accomplished using simple NumPy array slices --
# let's crop the face from the image
face = image[85:250, 85:220]

plt.title("Face")
plt.imshow(cv2.cvtColor(face, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# ...and now let's crop the entire body
body = image[90:450, 0:290]

plt.title("Body")
plt.imshow(cv2.cvtColor(body, cv2.COLOR_BGR2RGB))
plt.show()
```
