# Lighting and color spaces

```python
import numpy as np
import imutils
import cv2

# 資料視覺化模組
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 不要顯示惱人的 warning 訊息
import warnings
warnings.filterwarnings('ignore')

# 在 Notebook 裡可即時繪圖
%matplotlib inline  

plt.rcParams['figure.figsize'] = [15, 7]
plt.rcParams.update({'font.size':13})
```

## RGB

```python
# load the original image and show it (RGB)
image = cv2.imread("../input/florida_trip_small.png")
plt.title("RGB")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# loop over each of the individual channels and display them
for (name, chan) in zip(("B", "G", "R"), cv2.split(image)):
    plt.title(name)
    plt.imshow(chan, cmap="gray")
    plt.show()
```

## HSV

```python
# convert the image to the HSV color space and show it
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.title("HSV")
plt.imshow(hsv, cmap="hsv")
plt.show()
```

```python
# loop over each of the invidiaul channels and display them
for (name, chan) in zip(("H", "S", "V"), cv2.split(hsv)):
    plt.title(name)
    plt.imshow(chan, cmap="gray")
    plt.show()
```

## L\*a\*b\*

```python
# convert the image to the L*a*b* color space and show it
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
plt.title("L*a*b*")
plt.imshow(lab, cmap="jet")
plt.show()
```

```python
# loop over each of the invidiaul channels and display them
for (name, chan) in zip(("L*", "a*", "b*"), cv2.split(lab)):
    plt.title(name)
    plt.imshow(chan, cmap="gray")
    plt.show()
```

## Grayscale

```python
# show the original and grayscale versions of the image
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.title("Original")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

plt.title("Grayscale")
plt.imshow(gray, cmap="gray")
plt.show()
```
