# Smoothing and Blurring

```python
import numpy as np
import imutils
import cv2

# 
import matplotlib.pyplot as plt
import matplotlib.cm as cm

#  warning 
import warnings
warnings.filterwarnings('ignore')

#  Notebook 
%matplotlib inline  

# load the image and show it
image = cv2.imread("../input/florida_trip_small.png")

plt.rcParams['figure.figsize'] = [15, 7]
plt.rcParams.update({'font.size':13})
plt.title("Original")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

## Averaging

```python
kernelSizes = [(3, 3), (9, 9), (15, 15)]

# loop over the kernel sizes and apply an "average" blur to the image
for (kX, kY) in kernelSizes:
    blurred = cv2.blur(image, (kX, kY))
    plt.title("Average ({}, {})".format(kX, kY))
    plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
    plt.show()
```

## Gaussian

```python
# loop over the kernel sizes and apply a "Gaussian" blur to the image
for (kX, kY) in kernelSizes:
    blurred = cv2.GaussianBlur(image, (kX, kY), 0)
    plt.title("Gaussian ({}, {})".format(kX, kY))
    plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
    plt.show()
```

## Median

```python
# loop over the kernel sizes and apply a "Median" blur to the image
for k in (3, 9, 15):
    blurred = cv2.medianBlur(image, k)
    plt.title("Median {}".format(k))
    plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
    plt.show()
```
