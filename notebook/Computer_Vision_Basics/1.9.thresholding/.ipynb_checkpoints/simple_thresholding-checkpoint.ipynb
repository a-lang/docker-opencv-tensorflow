# Simple Thresholding

```python
import numpy as np
import imutils
import cv2

# 資料視覺化模組
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 不要顯示惱人的 warning 訊息
import warnings
warnings.filterwarnings('ignore')

# 在 Notebook 裡可即時繪圖
%matplotlib inline  

plt.rcParams['figure.figsize'] = [15, 7]
plt.rcParams.update({'font.size':13})
```

## Example: coins01.png

```python
# load the image, convert it to grayscale, and blur it slightly
image = cv2.imread("../input/coins01.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (7, 7), 0)
plt.title("Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# apply basic thresholding -- the first parameter is the image
# we want to threshold, the second value is is our threshold
# check; if a pixel value is greater than our threshold (in this
# case, 200), we set it to be BLACK, otherwise it is WHITE.
(T, threshInv) = cv2.threshold(blurred, 200, 255, cv2.THRESH_BINARY_INV)
plt.title("Threshold Binary Inverse")
plt.imshow(threshInv, cmap="gray")
plt.show()
```

```python
# using normal thresholding (rather than inverse thresholding),
# we can change the last argument in the function to make the coins
# black rather than white.
(T, thresh) = cv2.threshold(blurred, 200, 255, cv2.THRESH_BINARY)
plt.title("Threshold Binary")
plt.imshow(thresh, cmap="gray")
plt.show()
```

```python
# finally, we can visualize only the masked regions in the image
output = cv2.bitwise_and(image, image, mask=threshInv)
plt.title("Output")
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.show()
```
