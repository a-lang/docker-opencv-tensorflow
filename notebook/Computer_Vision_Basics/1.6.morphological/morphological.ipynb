# Morphological operations 

```python
import numpy as np
import imutils
import cv2

# 資料視覺化模組
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 不要顯示惱人的 warning 訊息
import warnings
warnings.filterwarnings('ignore')

# 在 Notebook 裡可即時繪圖
%matplotlib inline  

plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams.update({'font.size':13})
```

```python
# load the image and show it
image = cv2.imread("../input/pyimagesearch_logo.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.title("Original")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

## Erosion

```python
# apply a series of erosions
for i in range(0, 3):
    eroded = cv2.erode(gray.copy(), None, iterations=i + 1)
    plt.title("Eroded {} times".format(i + 1))
    plt.imshow(eroded, cmap="gray")
    plt.show()
```

## Dilation 

```python
# apply a series of dilations
for i in range(0, 3):
    dilated = cv2.dilate(gray.copy(), None, iterations=i + 1)
    plt.title("Dilated {} times".format(i + 1))
    plt.imshow(dilated, cmap="gray")
    plt.show()
```

## Opening 

```python
kernelSizes = [(3, 3), (5, 5), (7, 7)]

# load the image and show it
image2 = cv2.imread("../input/pyimagesearch_logo_noise.png")
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

plt.title("Original2")
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.show()

# loop over the kernels and apply an "opening" operation to the image
for kernelSize in kernelSizes:
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    opening = cv2.morphologyEx(gray2, cv2.MORPH_OPEN, kernel)
    plt.title("Opening: ({}, {})".format(kernelSize[0], kernelSize[1]))
    plt.imshow(opening, cmap="gray")
    plt.show()
```

## Closing 

```python
# loop over the kernels and apply a "closing" operation to the image
for kernelSize in kernelSizes:
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    closing = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel)
    plt.title("Closing: ({}, {})".format(kernelSize[0], kernelSize[1]))
    plt.imshow(closing, cmap="gray")
    plt.show()
```

## Morphological Gradient 

```python
# loop over the kernels and apply a "morphological gradient" operation
# to the image
for kernelSize in kernelSizes:
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    gradient = cv2.morphologyEx(gray, cv2.MORPH_GRADIENT, kernel)
    plt.title("Gradient: ({}, {})".format(kernelSize[0], kernelSize[1]))
    plt.imshow(gradient, cmap="gray")
    plt.show()
```
