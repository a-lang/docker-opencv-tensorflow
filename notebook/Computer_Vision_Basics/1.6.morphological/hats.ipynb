```python
import numpy as np
import imutils
import cv2

# 資料視覺化模組
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 不要顯示惱人的 warning 訊息
import warnings
warnings.filterwarnings('ignore')

# 在 Notebook 裡可即時繪圖
%matplotlib inline  

plt.rcParams['figure.figsize'] = [15, 7]
plt.rcParams.update({'font.size':13})
```

```python
# load the image and show it
image = cv2.imread("../input/car.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

## Top Hat/White Hat

```python
# construct a rectangular kernel and apply a blackhat operation which
# enables us to find dark regions on a light background
# NOTE: the kernel size (x, y) are odd integers
rectKernel = cv2.getStructuringElement(cv2.MORPH_RECT, (13, 5))
blackhat = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, rectKernel)

# similarly, a tophat (also called a "whitehat") operation will enable
# us to find light regions on a dark background
tophat = cv2.morphologyEx(gray, cv2.MORPH_TOPHAT, rectKernel)

# show the output images
plt.title("Original")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

plt.title("Blackhat")
plt.imshow(blackhat, cmap="gray")
plt.show()

plt.title("Tophat")
plt.imshow(tophat, cmap="gray")
plt.show()
```
