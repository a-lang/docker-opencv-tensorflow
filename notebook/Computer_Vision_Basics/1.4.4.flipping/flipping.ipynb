# Flipping

```python
import numpy as np
import imutils
import cv2

# 
import matplotlib.pyplot as plt

#  warning 
import warnings
warnings.filterwarnings('ignore')

#  Notebook 
%matplotlib inline  

# load the image and show it
image = cv2.imread("../input/florida_trip.png")

plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams.update({'font.size':13})
plt.title("Original")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# flip the image horizontally
flipped = cv2.flip(image, 1)

plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams.update({'font.size':13})
plt.title("Flipped Horizontally")
plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# flip the image vertically
flipped = cv2.flip(image, 0)

plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams.update({'font.size':13})
plt.title("Flipped Vertically")
plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# flip the image along both axes
flipped = cv2.flip(image, -1)

plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams.update({'font.size':13})
plt.title("Flipped Horizontally & Vertically")
plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
flipped_h = cv2.flip(image, 1)
(b, g, r) = flipped_h[235, 259]
print("R: {r}, G: {g}, B: {b}".format(b=b, g=g, r=r))
```

```python
rotated = imutils.rotate(flipped_h, 45)
flipped_v = cv2.flip(rotated, 0)
(b, g, r) = flipped_v[189, 441]
print("R: {r}, G: {g}, B: {b}".format(b=b, g=g, r=r))
```
