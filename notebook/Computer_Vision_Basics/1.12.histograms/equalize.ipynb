# Histograms Equalization 直方圖均化器

直方圖均化器可以藉由延伸圖素的分佈來改善影像的對比。

直方圖均化器要套用在灰階圖片。

> 當圖片的背景與前景都同為暗色或亮色時，這容易產生不真實的效果。在正常情況下，對衛星或醫學圖片增強對比度會很有用。


```python
import numpy as np
import imutils
import cv2

# 資料視覺化模組
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 不要顯示惱人的 warning 訊息
import warnings
warnings.filterwarnings('ignore')

# 在 Notebook 裡可即時繪圖
%matplotlib inline  

plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams.update({'font.size':13})
```

```python
# load the image and convert it to grayscale
image = cv2.imread("../input/beach.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# apply histogram equalization to stretch the constrast of our image
eq = cv2.equalizeHist(gray)

# show our images -- notice how the constrast of the second image has
# been stretched
plt.title("Original")
plt.imshow(gray, cmap="gray")
plt.show()

plt.title("Histogram Equalization")
plt.imshow(eq, cmap="gray")
plt.show()
```
