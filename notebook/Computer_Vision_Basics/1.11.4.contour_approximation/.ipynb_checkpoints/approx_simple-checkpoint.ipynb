# Contour approximation 近似法

```python
import numpy as np
import imutils
import cv2

# 資料視覺化模組
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 不要顯示惱人的 warning 訊息
import warnings
warnings.filterwarnings('ignore')

# 在 Notebook 裡可即時繪圖
%matplotlib inline  

plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams.update({'font.size':13})
```

## 找出方形物體

輪廓近似演算法使用
1. 先計算出輪廓的周長
2. 套用 cv2.approxPolyDP
3. 用 len() 得出連接點的總數，方形 = 4
4. 0.01，一般值範圍 1-5%，值愈大，計算出的連接點數就更少(意思是有更多的連接點被忽略)

```python
# load the the cirles and squares image and convert it to grayscale
image = cv2.imread("../input/circles_and_squares.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# find contours in the image
cnts = cv2.findContours(gray.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)

# loop over the contours
for c in cnts:
    # approximate the contour
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.01 * peri, True)

    # if the approximated contour has 4 vertices, then we are examining
    # a rectangle
    if len(approx) == 4:
        # draw the outline of the contour and draw the text on the image
        cv2.drawContours(image, [c], -1, (0, 255, 255), 2)
        (x, y, w, h) = cv2.boundingRect(approx)
        cv2.putText(image, "Rectangle", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX,
            0.5, (0, 255, 255), 2)

# show the output image
plt.title("Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# load the the cirles and squares image and convert it to grayscale
image = cv2.imread("../input/dog_contour.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# find contours in the image
cnts = cv2.findContours(gray.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)

# loop over the contours
for c in cnts:
    # approximate the contour
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.01 * peri, True)
    approx2 = cv2.approxPolyDP(c, 0.05 * peri, True)
    approx3 = cv2.approxPolyDP(c, 0.1 * peri, True)

print("length of approx(0.01) = {}".format(len(approx)))
print("length of approx(0.05) = {}".format(len(approx2)))
print("length of approx(0.1) = {}".format(len(approx3)))
```
