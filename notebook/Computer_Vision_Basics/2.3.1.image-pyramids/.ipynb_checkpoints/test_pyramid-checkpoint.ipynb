# Image pyramids

```python
from pyimagesearch.object_detection.helpers import pyramid
import cv2

# 資料視覺化模組
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 不要顯示惱人的 warning 訊息
import warnings
warnings.filterwarnings('ignore')

# 在 Notebook 裡可即時繪圖
%matplotlib inline  

#plt.rcParams['figure.figsize'] = [20, 10]
#plt.rcParams.update({'font.size':13})
```

```python
image = cv2.imread("./florida_trip.png")

# loop over the layers of the image pyramid and display them
for (i, layer) in enumerate(pyramid(image, scale=1.5)):
    # show the image
    plt.title("Layer {}".format(i + 1))
    plt.imshow(cv2.cvtColor(layer, cv2.COLOR_BGR2RGB))
    plt.show()
```
