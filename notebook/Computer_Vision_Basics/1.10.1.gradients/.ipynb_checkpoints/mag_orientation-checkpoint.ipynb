# Gradients
## Gradient orientation and magnitude in OpenCV

```python
import numpy as np
import imutils
import cv2

# 資料視覺化模組
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 不要顯示惱人的 warning 訊息
import warnings
warnings.filterwarnings('ignore')

# 在 Notebook 裡可即時繪圖
%matplotlib inline  

plt.rcParams['figure.figsize'] = [15, 7]
plt.rcParams.update({'font.size':13})
```

## Example: coins02.png

```python
# load the image, convert it to grayscale, and display the original
# image
image = cv2.imread("../input/coins02.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.title("Original")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
LOWER_ANGLE = 175.0
UPPER_ANGLE = 180.0

# compute gradients along the X and Y axis, respectively
gX = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
gY = cv2.Sobel(gray, cv2.CV_64F, 0, 1)

# compute the gradient magnitude and orientation, respectively
mag = np.sqrt((gX ** 2) + (gY ** 2))
orientation = np.arctan2(gY, gX) * (180 / np.pi) % 180

# find all pixels that are within the upper and low angle boundaries
idxs = np.where(orientation >= LOWER_ANGLE, orientation, -1)
idxs = np.where(orientation <= UPPER_ANGLE, idxs, -1)
mask = np.zeros(gray.shape, dtype="uint8")
mask[idxs > -1] = 255

# show the images
plt.title("Mask")
plt.imshow(mask, cmap="gray")
plt.show()
```
